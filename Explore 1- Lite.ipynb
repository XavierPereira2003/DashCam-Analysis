```python
from glob import glob
import re
import numpy as np
def convert_path(p):
    # Replace the folder part
    
    p = re.sub(r'^idd20kII/leftImg8bit/train', 'idd20kII/gtFine/train', p)
    # Replace the filename part
    p = re.sub(r'_leftImg8bit\.jpg$', '_gtFine_polygons.json', p)
    return p

image_paths = glob('idd20kII/leftImg8bit/train/*/*_leftImg8bit.jpg')
label_paths = [convert_path(p) for p in image_paths]

print(f"Length of the dataset: {len(image_paths)} and {len(label_paths)}" )
print(image_paths[0], label_paths[0])
```

```python
dataset=list(zip(image_paths,label_paths))
```

```python
import cv2
import json

colors = {
    'animal': (123, 113, 163),
    'autorickshaw': (132, 244, 127),
    'bicycle': (78, 10, 89),
    'billboard': (202, 234, 146),
    'bridge': (106, 83, 176),
    'building': (2, 187, 159),
    'bus': (179, 128, 239),
    'car': (54, 3, 145),
    'caravan': (22, 151, 149),
    'curb': (210, 67, 86),
    'drivable fallback': (45, 1, 55),
    'ego vehicle': (88, 229, 255),
    'fallback background': (187, 201, 150),
    'fence': (23, 245, 48),
    'ground': (251, 215, 108),
    'guard rail': (69, 206, 233),
    'license plate': (55, 78, 143),
    'motorcycle': (201, 103, 43),
    'non-drivable fallback': (18, 204, 64),
    'obs-str-bar-fallback': (128, 97, 141),
    'out of roi': (239, 34, 227),
    'parking': (84, 44, 78),
    'person': (17, 223, 133),
    'pole': (57, 121, 241),
    'polegroup': (136, 231, 194),
    'rail track': (120, 6, 75),
    'rectification border': (145, 94, 79),
    'rider': (163, 149, 253),
    'road': (85, 179, 52),
    'sidewalk': (6, 228, 159),
    'sky': (76, 62, 93),
    'traffic light': (154, 88, 174),
    'traffic sign': (92, 246, 147),
    'trailer': (53, 144, 121),
    'train': (165, 153, 30),
    'truck': (28, 91, 72),
    'tunnel': (10, 42, 84),
    'unlabeled': (24, 2, 34),
    'vegetation': (134, 4, 185),
    'vehicle fallback': (158, 91, 32),
    'wall': (242, 253, 127),
    'others': (0, 0, 0)
}
```

```python
for image_path, label_path in dataset[:4]:
    with open(label_path, 'r') as f:
        data = json.load(f)
    print(image_path)
    image = cv2.imread(image_path)

    imgHeight, imgWidth = data.get('imgHeight'), data.get('imgWidth')
    for obj in data.get('objects'):
        label = obj.get('label')
        polygon = obj.get('polygon')

        pts = np.array(polygon, dtype=np.float32)
        pts = np.round(pts)
        pts = np.array(pts, dtype=np.int32)
        pts = pts.reshape((-1, 1, 2))
        try:
            color = colors[label]
        except KeyError:
            color = colors['others']

        overlay = image.copy()

        alpha=0.5
        cv2.polylines(overlay, [pts], isClosed=True, color=color, thickness=1)
        image = cv2.addWeighted(overlay, alpha, image, 1 - alpha, 0)

        x, y, w, h = cv2.boundingRect(pts)
        cv2.putText(image, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    # Add legends
    legend_x, legend_y = 10, 30
    for label, color in colors.items():
        cv2.rectangle(image, (legend_x, legend_y), (legend_x + 20, legend_y + 20), color, -1)
        cv2.putText(image, label, (legend_x + 30, legend_y + 15), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
        legend_y += 30

    # Display the annotated image.
    cv2.imshow("Annotated Image", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
